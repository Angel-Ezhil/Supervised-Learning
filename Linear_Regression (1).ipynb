```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


d=pd.read_csv("data (1).csv")
print(d.loc[2:16,["Height","Weight"]])
x=d[["Height"]]
y=d[["Weight"]]


model = LinearRegression()
model.fit(x, y)


y_pred = model.predict(x)

r2 = r2_score(y, y_pred)

print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
print("R2 Score:", r2)

# Predict for 1.70 Height
prediction = model.predict(pd.DataFrame([[1.7]], columns=['Height']))
print("Predicted Weight for 1.70 Hours:",prediction[0])


sns.scatterplot(x=x.values.flatten(), y=y.values.flatten(), label="Actual Data")
sns.lineplot(x=x.values.flatten(), y=y_pred.flatten(), label="Regression Line")

plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight Regression")
plt.show()
```
